# One-Hot Encoding
Perform one-hot encoding to convert items into binary columns. Each column represents an item, and a '1' in a column indicates that the item was purchased in the transaction, while '0' indicates it was not.

```python
import pandas as pd

# Load the dataset from the Excel file
dataset_path = '/kaggle/input/market-basket-analysis/Assignment-1_Data.xlsx'
df = pd.read_excel(dataset_path)

# Split the 'Itemname' column into lists of items
df['Itemname'] = df['Itemname'].str.split(', ')

# Perform one-hot encoding
one_hot = pd.get_dummies(df['Itemname'].apply(pd.Series).stack()).sum()

# Concatenate the one-hot encoded DataFrame with the original
result = pd.concat([df['BillNo'], one_hot], axis=1)

# Fill NaN values with 0
result = result.fillna(0)

print(result)
```

# Association Analysis
With the preprocessed data, you can perform association analysis to find itemsets that are frequently bought together. This can be done using algorithms like Apriori or FP-growth.


```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataset_path = '/kaggle/input/market-basket-analysis/Assignment-1_Data.xlsx'
df = pd.read_excel(dataset_path)

# Convert the data into a one-hot encoded format
basket = pd.get_dummies(df.set_index('BillNo')).astype(bool)

# Find frequent item sets using Apriori
frequent_itemsets = apriori(basket, min_support=0.5, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Display frequent item sets and association rules
print("Frequent Item Sets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```

# Association Rule Mining
Use association rule mining algorithms like Apriori or FP-growth to generate association rules from the transaction data. Python libraries like mlxtend are commonly used for this purpose.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Load your market basket dataset
dataset_path = '/kaggle/input/market-basket-analysis/Assignment-1_Data.xlsx'
df = pd.read_excel(dataset_path)

# Assuming your dataset has a column named 'Transaction' containing transaction IDs
# and a column named 'Items' containing a list of items in each transaction

# Split the 'Items' column into lists of items
df['Itemname'] = df['Itemname'].str.split(', ')

# Convert the data into a binary format for Apriori
oht = pd.get_dummies(df['Itemname'].apply(pd.Series).stack())

# Use Apriori to find frequent itemsets
frequent_itemsets = apriori(oht, min_support=0.001, use_colnames=True)

# Use Association Rules to find interesting associations
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Sort the rules by confidence in descending order
rules = rules.sort_values(by='confidence', ascending=False)
# Display the results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```
